```python
People = {
    'first' : ['Rohit','Avi'],
    'Last'  : ['Ratawal','Gupta'],
    'Language' : ['Python','Vba']
}
```

```python
import pandas as pd

df = pd.DataFrame(People)
```

```python
df
```

```python
df =df.append({'first':'Ravi','Last':'Sharma','Language':'Sql'},ignore_index=True)
```

```python
df.rename(columns={'first':'First'})
```

```python
df.columns = df.columns.str.upper()
df
```

```python
df.columns = df.columns.str.lower()
df
```

```python
#Filter and remove
flt = df['first'] == 'Ravi'
df = df.drop(index=df[flt].index)
df
```

```python
df.insert(2,'Full Name', df['first'] + ' ' + df['last'])
df
```

```python
df.columns = df.columns.str.replace(' ','_')
df
```

```python
df = df.drop(columns='Full_Name')
df
```

```python
df.sort_values('first', ascending = True)
```

```python
df.sort_values('first', ascending = False)
```

```python
# sorting on any specific column
df.sort_values(by=['first','last'], ascending=[False,True])
```

```python
df.sort_index()
```
