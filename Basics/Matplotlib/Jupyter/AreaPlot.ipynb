```python
%matplotlib notebook
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd

pd.set_option('display.max_columns', 500)
data_c = []
data_c = pd.read_excel('/home/rohyth/Downloads/UN_MigFlow_All_CountryFiles/Canada.xlsx',sheet_name='Canada by Citizenship')
data_c.tail()

data_c.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
data_c.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
data_c.set_index('Country', inplace=True)
data_c['Total'] = data_c.sum(axis=1)
```

```python
data_c.sort_values(['Total'], ascending=False, axis=0, inplace=True)

d_top5  = data_c.loc[['India','China','United Kingdom of Great Britain and Northern Ireland','Philippines','Pakistan']]

years = [x for x in range(1980,2013)]
d_top5 = d_top5[years].transpose()
d_top5.head()
```

```python
mpl.style.use('seaborn')
d_top5.plot(kind='area')

plt.title('Top 5 Countries')
plt.xlabel('years')
plt.ylabel('Immigrants')
plt.tight_layout()
plt.show()
```
