```python
%matplotlib notebook
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd

pd.set_option('display.max_columns', 500)
data_c = pd.read_excel('/home/rohyth/Downloads/UN_MigFlow_All_CountryFiles/Canada.xlsx',sheet_name='Canada by Citizenship',index_col='OdName')
data_c.head()
```

```python
years = [x for x in range(1980,2013)]
mpl.style.use(['ggplot'])
data_c.loc['Haiti',years].plot(kind='line')
plt.text(2000, 6000, '2010 Earthquake')
plt.xlabel('Years')
plt.ylabel('Immigrants')
plt.title('Immigration Status for Haiti')
plt.tight_layout()
plt.show()
```

```python
mpl.style.use('seaborn')
data_c.loc['India',years].plot(kind='line')
data_c.loc['China',years].plot(kind='line')

plt.legend()
plt.xlabel('Years')
plt.ylabel('Immigrants')
plt.title('Immigration Status for Haiti')
plt.tight_layout()
plt.show()
```

```python
mpl.style.use('seaborn')
d_CI = data_c.loc[['India','China'],years]
d_CI = d_CI.transpose()
d_CI.head()
```

```python
d_CI.plot(kind='line')

plt.legend()
plt.xlabel('Years')
plt.ylabel('Immigrants')
plt.title('Immigration Status for Haiti')
plt.tight_layout()
plt.show()
```
